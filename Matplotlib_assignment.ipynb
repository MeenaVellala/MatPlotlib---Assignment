Analysis:
Below are the study analysis results from Pymaceuticals' research on 249 mice diagnosed with Squamous Cell Carcinoma (SCC) tumors and treated with various drug regimens. We identified and removed one duplicate sample with conflicting data during our review. The final data analysis includes:

Summary statistics of central tendency, variance, and standard error of tumor volume for each drug regimen.
Total time points measured for all mice per drug regimen.
Comparison of male-to-female distribution among the test subjects.
Distribution of final tumor volumes across drug regimens, highlighting one outlier.
Tumor volume over time for mice treated with Capomulin (specifically, mouse l509).
Tumor volume versus mouse weight for the Capomulin treatment regimen, including a linear regression model and correlation coefficient.
Based on the summary statistics table, the mean tumor volume for the drugs Capomulin and Ramicane is lower compared to the placebo. A bar plot was generated showing the total number of mice used within each drug regimen. The drug regimens Capomulin and Ramicane had over 200 mice within the trial studies, while all other drug regimens had fewer than 200 mice.

The sex ratio of mice used is 51% males and 49% females, indicating that the studies were balanced in terms of potential sex differences in reaction to the drug regimens.

Further analysis was completed to identify the final tumor volume across four drug regimens: Capomulin, Ramicane, Infubinol, and Ceftamin. The distribution of tumor volume for these four treatment groups varied. Capomulin and Ramicane had smaller final tumor volumes, with their average final tumor size under 40mm³. Conversely, the drug regimens Infubinol and Ceftamin had larger final tumor volumes, with an average size around 60mm³.

Tumor volume over time was compared for a single mouse (l509), indicating a decrease in tumor size over time. However, as this analysis was based on one animal, it does not necessarily indicate that Capomulin had this effect on all mice within the trial.

A scatter plot was created to compare the average tumor size to the average weight (g) of the animals within the Capomulin drug regimen trials. The data showed a correlation coefficient of 0.84 and a linear regression equation of y = 0.95x + 21.55. These results indicate a positive correlation between the animals' weight and their final tumor volume within the Capomulin drug trials omuline 



```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
from scipy.stats import linregress

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data 
mouse_metadata = pd.read_csv(mouse_metadata_path)
mouse_metadata.head()


```

```python
# Read the study results
study_results = pd.read_csv(study_results_path)
study_results.head()
```

```python
# Combine the data into a single DataFrame
mouse_df = pd.merge(study_results,mouse_metadata, on="Mouse ID")
# Display the data table for preview
mouse_df.head()
```

```python
# Checking the number of mice.
mice =mouse_df["Mouse ID"].value_counts()
mice_count= len(mice)
mice_count
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint.
mouse_dup = mouse_df.loc[mouse_df.duplicated(subset=['Mouse ID', 'Timepoint']),'Mouse ID'].unique()
mouse_dup
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
dup_mouse_df = mouse_df.loc[mouse_df["Mouse ID"] == "g989", :]
dup_mouse_df
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
clean_dup_df = mouse_df[mouse_df["Mouse ID"].isin(dup_mouse_df)==False]
# Identify rows present in the second DataFramei.e., duplicate data frame
mask = mouse_df.isin(dup_mouse_df.to_dict('list')).all(axis=1)
# Filter out these rows
clean_df = mouse_df[~mask]
clean_df.head()
```

```python
# Checking the number of mice in the clean DataFrame.
mice_clean=clean_df["Mouse ID"].value_counts()
mice_clean_count=len(mice_clean)
mice_clean_count
```

Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen:
# Group by `drug regimen` and take the mean of the `Tumor Volume` column for each.
Mean_Tumor_Volume = clean_df.groupby("Drug Regimen").mean(numeric_only=True)["Tumor Volume (mm3)"]
#Mean_Tumor_Volume
Median_Tumor_Volume = clean_df.groupby("Drug Regimen").median(numeric_only=True)["Tumor Volume (mm3)"]
#Median_Tumor_Volume
Variance_Tumor_Volume = clean_df.groupby("Drug Regimen").var(numeric_only=True)["Tumor Volume (mm3)"]
#Variance_Tumor_Volume
Std_Tumor_Volume = clean_df.groupby("Drug Regimen").std(numeric_only=True)["Tumor Volume (mm3)"]
#Std_Tumor_Volume
Sem_Tumor_Volume = clean_df.groupby("Drug Regimen").sem(numeric_only=True)["Tumor Volume (mm3)"]
#Sem_Tumor_Volume
# mean, median, variance, standard deviation, and SEM of the tumor volume.
# Assemble the resulting series into a single summary DataFrame.
stat_summary = pd.DataFrame({"Mean Tumor Volume": Mean_Tumor_Volume, "Median Tumor Volume": Median_Tumor_Volume,
                             "Tumor Volume Variance": Variance_Tumor_Volume, "Tumor Volume Std. Dev.": Std_Tumor_Volume, 
                             "Tumor Volume Std. Err.": Sem_Tumor_Volume})
stat_summary
```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)

# Using the aggregation method, produce the same summary statistics in a single line
summary_agg =  clean_df.groupby(['Drug Regimen'])[['Tumor Volume (mm3)']].agg(['mean', 'median', 'var', 'std', 'sem'])
summary_agg
```

Quartiles, Outliers and Boxplots


```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse
max_timepoint_df = pd.DataFrame((clean_df.groupby(["Mouse ID"])["Timepoint"]).max().reset_index())
max_timepoint_df 
# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
merged_max_df=  max_timepoint_df.merge(clean_df,on=['Mouse ID', 'Timepoint'], how="left")
merged_max_df
```

```python
# Put treatments into a list for for loop (and later for plot labels)
treatment_group=["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]

# Create empty list to fill with tumor vol data (for plotting)
tumor_vol_data=[]

# Calculate the IQR and quantitatively determine if there are any potential outliers
# Iterate through each drug regimen
for t in range(len(treatment_group)): 

    # Locate the rows which contain mice on each drug and get the tumor volumes
    tumor = merged_max_df.loc[merged_max_df['Drug Regimen'] == treatment_group[t], ['Tumor Volume (mm3)']]
    
    # add subset 
    tumor_vol_data.append(tumor['Tumor Volume (mm3)'])
    
    # Determine outliers using upper and lower bounds
    # Calculate IQR to find outliers
    quartiles = tumor_vol_data[t].quantile([.25, .5, .75])
    lowrq = quartiles[.25]
    upprq = quartiles[.75]
    iqr = upprq - lowrq
    lwr_bnd = lowrq - (1.5 * iqr)
    upr_bnd = upprq + (1.5 * iqr)

    # Search for and print outliers
    out = tumor_vol_data[t].loc[(tumor_vol_data[t] < lwr_bnd) | (tumor_vol_data[t] > upr_bnd)]
    print(f"{treatment_group[t]}'s potential outliers: {out}")
```

```python
# Generate a box plot that shows the distribution of the tumor volume for each treatment group.
flierprops = dict(marker = 'o', markerfacecolor = 'r', markeredgecolor = 'black' )

# create boxplot -- include x-axis labels, add flierprops parameter to highlight outlier
plt.boxplot(tumor_vol_data, labels = treatment_group, flierprops = flierprops)

# add label to y-axis 

plt.ylabel("Final Tumor Volume (mm3)")

# display boxplot
plt.show()
```

Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a single mouse treated with Capomulin

# find the list of mice treated with Capomulin, to identify possible mice to graph in line plot
Capomulin = clean_df.loc[clean_df["Drug Regimen"] == "Capomulin"]["Mouse ID"].unique()
#Capomulin
# filter dataframe for mouse l509, to return series with timepoint and tumor volume size
sample_mouse = clean_df.loc[clean_df["Mouse ID"] == "l509",["Timepoint","Tumor Volume (mm3)"]]

# store values for timepoint and tumor volumn in variables to plot in x and y axis
x = sample_mouse["Timepoint"]
y = sample_mouse["Tumor Volume (mm3)"]

# set figure size, create line plot showing time and tumor volume for the sample mouse
plt.figure(figsize = (6,4))
plt.plot(x,y)

# display chart with axis labels and chart title
plt.xlabel("Timepoint (days)")
plt.ylabel("Tumor Volume (mm3)")
plt.title("Capomulin treatment of mouse l509")

plt.show()
```

```python
# Generate a scatter plot of average tumor volume vs. mouse weight for the Capomulin regimen

# create a dataframe for mice treated ONLY with the Capomulin regimen
capomulin_df = pd.DataFrame(clean_df.loc[clean_df["Drug Regimen"] == "Capomulin",:])
#capomulin_df

# calculate the average tumor volume per mouse, and use a variable to store values
avg_tumor_volume = capomulin_df.groupby(["Mouse ID"])["Tumor Volume (mm3)"].mean()
#avg_tumor_volume 
# calculate the average weight per mouse, and use a variable to store values
avg_weight = capomulin_df.groupby(["Mouse ID"])["Weight (g)"].mean()
#avg_weight
# create Data Frame with average tumor, average weight 
capomulin_average = pd.DataFrame({"Avg. Tumor": avg_tumor_volume, "Avg. Weight": avg_weight})
#capomulin_average
# add scatter plot showing average tumor volume and average weight
x = avg_weight
y = avg_tumor_volume

# set figure size, create scatter plot
plt.figure(figsize = (6,4))
plt.scatter(x, y)

# display chart with axis labels and chart title
plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")
plt.title("Capomulin Mice - Average Tumor Weight & Volume")
plt.show()
```

Correlation and Regression

```python
# Calculate the correlation coefficient and a linear regression model
# for mouse weight and average observed tumor volume for the entire Capomulin regimen
correlation = st.pearsonr(x,y)
print(f"The correlation between mouse weight and the average tumor volume is {round(correlation[0],2)}.")
# calculate the linear regression
(slope, intercept, rvalue, pvalue, stderr) = linregress(x,y)

regression_equation = x * slope + intercept

line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))

plt.scatter(x, y)

plt.plot(x, regression_equation, "r-")

plt.annotate(line_eq,(17,37),fontsize=15,color="red")

# display chart with axis labels and chart title
plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")
plt.show()
```
